## Test results log (progress)

**How this works:** Run the Python cells below. When you **Save** the notebook, Jupyter stores **outputs** inside this `.ipynb` file — so you get a lightweight, commit-friendly history of “what passed when.”

- Prefer committing after a green run (or after a deliberate “red” snapshot you care about).
- **Vitest** is fast; **E2E** needs Playwright browsers installed (`npx playwright install chromium`).
- If the kernel’s cwd is wrong, cells resolve the repo root automatically (`package.json` next to `src/`).


### Vitest (unit / RTL)


````python
# Run Vitest; output is captured below when you execute + save the notebook.
import subprocess
import sys
from pathlib import Path
from datetime import datetime, timezone

def repo_root() -> Path:
    here = Path.cwd().resolve()
    if (here / "package.json").exists():
        return here
    if (here.parent / "package.json").exists():
        return here.parent
    raise FileNotFoundError("Start Jupyter from repo root or from docs/ — could not find package.json")

ROOT = repo_root()
stamp = datetime.now(timezone.utc).strftime("%Y-%m-%d %H:%M:%S UTC")
print(f"=== Vitest @ {stamp} ===")
print(f"repo: {ROOT}\n")

proc = subprocess.run(
    ["npm", "run", "test", "--", "--reporter=verbose"],
    cwd=ROOT,
    capture_output=True,
    text=True,
)
sys.stdout.write(proc.stdout or "")
if proc.stderr:
    sys.stdout.write("\n--- stderr ---\n")
    sys.stdout.write(proc.stderr)
print(f"\n[exit code {proc.returncode}]")
if proc.returncode != 0:
    raise RuntimeError(f"Vitest failed with exit code {proc.returncode}")
````

### Playwright E2E (slower)

Runs `npm run test:e2e` (local config builds + preview unless `CI=true` is set). Set **`CI=1`** in the environment *before* running this cell to match CI (run `npm run build` first in a terminal, then `CI=1 npm run test:e2e`).


````python
import os
import subprocess
import sys
from pathlib import Path
from datetime import datetime, timezone

def repo_root() -> Path:
    here = Path.cwd().resolve()
    if (here / "package.json").exists():
        return here
    if (here.parent / "package.json").exists():
        return here.parent
    raise FileNotFoundError("Could not find package.json")

ROOT = repo_root()
stamp = datetime.now(timezone.utc).strftime("%Y-%m-%d %H:%M:%S UTC")
print(f"=== Playwright E2E @ {stamp} ===")
print(f"repo: {ROOT}")
print(f"CI env: {os.environ.get('CI', '')!r}\n")

env = os.environ.copy()
proc = subprocess.run(
    ["npm", "run", "test:e2e"],
    cwd=ROOT,
    capture_output=True,
    text=True,
    env=env,
)
sys.stdout.write(proc.stdout or "")
if proc.stderr:
    sys.stdout.write("\n--- stderr ---\n")
    sys.stdout.write(proc.stderr)
print(f"\n[exit code {proc.returncode}]")
if proc.returncode != 0:
    raise RuntimeError(f"Playwright failed with exit code {proc.returncode}")
````
